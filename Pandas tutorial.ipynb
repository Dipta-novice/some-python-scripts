### Distance between two euclidian points

```python
temp_df=pd.DataFrame(
    {'point1':[(1,2),(2,4)],
     'point2':[(3,4),(5,6)]
    }
)
temp_df
```

```python
def euclidian(row):
    pt_A=row['point1']
    pt_B=row['point2']
    distance=((pt_A[0]-pt_B[0])**2+(pt_A[1]-pt_B[1])**2)**0.5
    return(distance)
```

```python
temp_df['Euclidian_distance']=temp_df.apply(euclidian,axis=1)
temp_df
```

```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('imdb-top-1000.csv')
movies.head(2)
```

```python
movies.info()
```

```python
movies.groupby('Genre')[['Runtime','IMDB_Rating','No_of_Votes','Gross','Metascore']].sum()
```

```python
# find the top 3 genres by total earning
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).nlargest(3)
```

```python
# find the genre with highest avg IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

['IMDB_Rating'].idxmax():

Within each genre group, finds the index of the row where the IMDB_Rating is the highest.
Returns a Series of indices, one for each genre, corresponding to the row with the maximum IMDb rating.

```python
# find the highest rated movie of each genre
movies.loc[movies.groupby('Genre')['IMDB_Rating'].idxmax(), ['Genre', 'Series_Title', 'IMDB_Rating']]
```

```python
movies.groupby('Genre').agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list
movies.groupby('Genre').agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```
