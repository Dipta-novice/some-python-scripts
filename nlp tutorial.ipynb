Tokenization

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# Ensure NLTK resources are downloaded
nltk.download('punkt')
nltk.download('punkt_tab')
# Example text
text = "Tokenization is the process of splitting text into meaningful units, such as words or sentences."

# Word tokenization
word_tokens = word_tokenize(text)
print("Word Tokens:", word_tokens)

# Sentence tokenization
sentence_tokens = sent_tokenize(text)
print("Sentence Tokens:", sentence_tokens)
```

Finding frequency 

```python
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
import nltk

# Ensure NLTK resources are downloaded
nltk.download('punkt')

# Sample text
text = "Tokenization is the process of splitting text into meaningful units. Tokenization is important in NLP."

# Tokenize the text into words
tokens = word_tokenize(text)

# Create a frequency distribution
freq_dist = FreqDist(tokens)

# Display the frequencies of each word
print("Word Frequencies:")
for word, freq in freq_dist.items():
    print(f"{word}: {freq}")

# Plot the frequencies
print("\nPlotting word frequencies...")
freq_dist.plot(10, title="Word Frequency Distribution")
```

Stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import nltk

# Ensure NLTK resources are downloaded
nltk.download('punkt')

# Sample text
text = "The cats are playing with the balls. They played happily yesterday."

# Tokenize the text into words
tokens = word_tokenize(text)

# Initialize the PorterStemmer
stemmer = PorterStemmer()

# Apply stemming to each word
stemmed_words = [stemmer.stem(word) for word in tokens]

# Display original and stemmed words
print("Original Words:", tokens)
print("Stemmed Words:", stemmed_words)
```

lemmatization

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk

# Ensure NLTK resources are downloaded
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Sample text
text = "The cats are playing with the balls. They played happily yesterday."

# Tokenize the text into words
tokens = word_tokenize(text)

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Apply lemmatization to each word (using 'v' for verb context and 'n' for noun context)
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in tokens]  # for verbs
# You can also use pos='n' for nouns if you want to handle nouns separately

# Display original and lemmatized words
print("Original Words:", tokens)
print("Lemmatized Words (Verbs):", lemmatized_words)
```

Stop words removal

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Ensure NLTK resources are downloaded
nltk.download('punkt')
nltk.download('stopwords')

# Sample text
text = "This is a sample sentence, and we are removing stopwords from it."

# Tokenize the text into words
tokens = word_tokenize(text)

# Get the list of English stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords from the tokenized words
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

# Display original and filtered tokens
print("Original Tokens:", tokens)
print("Filtered Tokens (without stopwords):", filtered_tokens)
```

POS Tagging

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

# Make sure to download the necessary NLTK resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Sample sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Tokenize the sentence into words
words = word_tokenize(sentence)

# Perform POS tagging
pos_tags = pos_tag(words)

# Print the POS tags
print(pos_tags)
```

Bag of Words


```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample text corpus
corpus = [
    "I love programming in Python",
    "Python programming is fun",
    "I love creating machine learning models",
    "Machine learning is fascinating"
]

# Initialize CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the corpus
X = vectorizer.fit_transform(corpus)

# Get feature names
feature_names = vectorizer.get_feature_names_out()

# Convert the sparse matrix to a dense array for visualization
dense_matrix = X.toarray()

# Display the Bag of Words representation
print("Feature Names:", feature_names)
print("Dense Matrix:")
print(dense_matrix)
```

Cosine similarity

```python
import spacy
from spacy import displacy
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load the SpaCy model
nlp = spacy.load('en_core_web_sm')

# Sample texts
text1 = "Apple is looking at buying U.K. startup for $1 billion."
text2 = "San Francisco is a beautiful city and Google is planning to expand its headquarters."

# Process the texts with the SpaCy model
doc1 = nlp(text1)
doc2 = nlp(text2)

# Get the vector representation of the documents
vector1 = doc1.vector
vector2 = doc2.vector

# Compute cosine similarity using sklearn
cos_sim = cosine_similarity([vector1], [vector2])

# Print the cosine similarity
print(f"Cosine Similarity: {cos_sim[0][0]}")
```

```python
import gensim.downloader as api

# Load a pre-trained Word2Vec model
model = api.load("word2vec-google-news-300")  # This is a pre-trained Word2Vec model on Google News dataset

# Example words
word1 = "king"
word2 = "queen"

# Get the vector representation of the words
vector_word1 = model[word1]
vector_word2 = model[word2]

print(f"Vector for '{word1}':\n{vector_word1}")
print(f"Vector for '{word2}':\n{vector_word2}")

# Cosine similarity between the two words
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity([vector_word1], [vector_word2])
print(f"Cosine similarity between '{word1}' and '{word2}': {similarity[0][0]}")
```

TF IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text corpus
corpus = [
    "I love programming in Python",
    "Python programming is fun",
    "I love creating machine learning models",
    "Machine learning is fascinating"
]

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the corpus
X = vectorizer.fit_transform(corpus)

# Get feature names
feature_names = vectorizer.get_feature_names_out()

# Convert the sparse matrix to a dense array for visualization
dense_matrix = X.toarray()

# Display the TF-IDF representation
print("Feature Names:", feature_names)
print("Dense Matrix:")
print(dense_matrix)
```

N grams


```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample text corpus
corpus = [
    "I love programming in Python",
    "Python programming is fun",
    "I love creating machine learning models",
    "Machine learning is fascinating"
]

# Initialize CountVectorizer with n-gram range (bigrams and trigrams)
vectorizer = CountVectorizer(ngram_range=(2, 3))  # Change (2, 3) for different n-gram ranges

# Fit and transform the corpus
X = vectorizer.fit_transform(corpus)

# Get feature names
feature_names = vectorizer.get_feature_names_out()

# Convert the sparse matrix to a dense array for visualization
dense_matrix = X.toarray()

# Display the n-grams and their counts
print("N-Gram Feature Names:")
print(feature_names)
print("\nDense Matrix:")
print(dense_matrix)
```
